```python
# Pie chart.

plt.figure(figsize=(8,8))
plt.pie(Merchant['Amount (with decimal mark per currency exponent)'], labels=Merchant['Type'], autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Merchant Status')
plt.show()
```

```python
import geopandas as gpd

world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
amount = Merchant.groupby('BIN country')['Amount (with decimal mark per currency exponent)'].sum().reset_index()
world = world.merge(amount, left_on='name', right_on='BIN country', how='left')

fig, ax = plt.subplots(1,1, figsize=(15,10))
world.plot(column='Amount (with decimal mark per currency exponent)', ax=ax, legend=True, legend_kwds={'label':'Amount by Country'})
ax.set_title('Amount distribution')
plt.show()
```

```python
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

world = world.merge(Merchant, left_on='name', right_on='BIN country', how='left')

# Plot the world map
world.plot(column='Amount (with decimal mark per currency exponent)', cmap='OrRd', legend=True, legend_kwds={'label': "Total Sales by Country"})
plt.show()
```
