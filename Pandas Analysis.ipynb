```python
# Date 5 Feb 2024.
# Problem : Analysing Merchant data, graphically using only pandas.
# Documentation & solution by Akhona Njeje.

import pandas as pd
import numpy as np

# dummy/test data.

data = {
    'Name':['Ak', 'XY', 'x', 'Art', 'Real'],
    'Age':[25, 30, 20, 40, 15],
    'Salary':[100, 200, 10, 1000, 5]
}
df = pd.DataFrame(data)
df
```

```python
# Bar.

df.plot(x='Name', y='Salary', kind='bar', title='Salary Distribution')
```

```python
# Pie

df.plot.pie(y='Age', labels=df['Name'], autopct='%1.1f%%', title='Age Distribution')
```

```python
# Horizontal bar.

df.plot(x='Name', y='Salary', kind='barh', title='Salary Distribution')
```

```python
# Line.

df.plot(x='Name', y=['Age', 'Salary'], kind='line', title='Salary Distribution')
```

```python
df.plot.scatter(x='Age', y='Salary', title='Age vs Salary')
```

```python
# KPI's.

average_age = df['Age'].mean()
average_sal = df['Salary'].mean()
total_records = len(df)
highest_salary = df['Salary'].max()
```

```python
# Create KPI's.

kpi_data = {
    'Metrics':['Ave age', 'Ave salary', 'Total records', 'Highest salary'],
    'Value':[average_age, average_sal, total_records, highest_salary]
}

kpi_df = pd.DataFrame(kpi_data)
print("KPI's:")
print(kpi_df)
```
